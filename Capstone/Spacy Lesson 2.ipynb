```python
# Word Counts
df["words"].apply(lambda x: len(str(x).split()))
```

```python
#Lowercase conversion
df["words"] = df["words"].apply(lambda x: str(x).lower())
```

```python
#Special Characters Removal

# re.sub(r'[^\w]+',"",string=string)
```

```python
re.sub(r'[^\w]+',"",string=test)
```

```python
df["words"] = df["words"].apply(lambda x: re.sub(r'[^\w]+',"",string=x))
```

```python
# Remove multiple spaces
```

```python
# ' '.join(x.split())
```

```python
df["words"] = df["words"].apply(lambda x: ' '.join(x.split()))
```

```python
# Remove stop words

# ' '.join([t for t in x.split() if t not in stopwords])
```

```python
df["words"] = df["words"].apply(lambda x: ' '.join([t for t in x.split() if t not in stopwords]))
```
